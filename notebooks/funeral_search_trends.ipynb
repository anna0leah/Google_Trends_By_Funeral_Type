```python
import pandas as pd
```

```python
## open csvs
funeral_term=pd.read_csv("../data/funeral_term_subregion.csv")
term_funeral=pd.read_csv("../data/term_funeral_subregion.csv")
```

```python
funeral_term.head()
```

```python
term_funeral.head()
```

```python
## append csvs
searches = pd.merge(term_funeral, funeral_term, on="State", how="left")
searches.head()
```

```python
len(searches)
```

```python
type("funeral_for_miscarriage")
```

```python
##convert to int
searches.astype({"miscarried_funeral":"int", "stillbirth_funeral":"int", "stillborn_funeral":"int", "funeral_for_stillborn":"int", "funeral_for_miscarriage":"int"}) 
```

```python
## combine similar columns
searches["miscarriage"] = searches.apply(lambda x: x["miscarriage_funeral"]+ x["miscarried_funeral"] + x["funeral_for_miscarriage"], axis=1)
searches["stillbirth"] = searches.apply(lambda x: x["stillbirth_funeral"]+ x["stillborn_funeral"] + x["funeral_for_stillborn"], axis=1)
```

```python
searches.head()
```

```python
compare_searches=searches[["State","miscarriage","stillbirth"]]
```

```python
compare_searches.head()
```

```python
## create and export map
compare_searches.to_csv("../output/miscarriage_stillbirth_searches.csv", index=False)
```
